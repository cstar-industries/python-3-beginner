## 2.

Extend the class from the previous exercise, so that the instances can greet more then just the world.

The `greet()` method should be rewritten as `greet(target)` and output `Hello <target>!`

Usage example:

```python
g = Greeter()
g.greet('World')
g.greet('Python')
```
Output:
```
Hello World!
Hello Python!
```

> **Optional:** In a real code situation, you'd want the new `Greeter` not to break previous usage. Use a _default argument_ to make sure that the `greet()` case still outputs `Hello World!`. After redefining your `Greeter` class, try running the previous test blocks again, and see if you still get the expected output.

````
class Greeter:
  def greet(self, target='World'):
    print(f'Hello {target}!')
````

````
# Test block

g = Greeter()
g.greet('World')
g.greet('Python')
````

## 3.

Extend the class from the previous exercise to give our greeter their own personality.

The `Greeter` class should now have its own greeting phrase. The `Greeter` constructor should take a phrase as argument. Calling `greet` should replace `Hello` with the greeting.

Usage example:

```python
g = Greeter('Hello')
g.greet('World')
g = Greeter('Howdy')
g.greet('Python')
```
Output:
```
Hello World!
Howdy Python!
```

> **Optional:** Once again, you can make good use of default arguments and uphold backwards compatibility with previous implementations. Try re-running the test blocks from exercise 1. and 2.

````
class Greeter:
  def __init__(self, greeting='Hello'):
    self.greeting = greeting
    
  def greet(self, target='World'):
    print(f'{self.greeting} {target}!')
````

````
g = Greeter('Hello')
g.greet('World')

g = Greeter('Howdy')
g.greet('Python')
````
