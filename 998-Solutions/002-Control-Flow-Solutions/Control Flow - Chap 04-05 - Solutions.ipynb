## 1.

Add all the numbers in the list `sizes` from left to right until you go over the `limit`. Print the sum you reached. Use a `while` loop to check for your exit condition.

Example:

```python
limit = 40
sizes = [1, 3, 33, 12, 9, 41, 22]
```

should output `49` (`1 + 3 + 33 + 12`).

````
limit = 3300
sizes = [851,431,901,486,536,978,214,101,404,557,548,1000,17,654,881,893,486,514,741,63]

total = 0
i = 0
while total < limit:
  total += sizes[i]
  i += 1
print(total)
````

## 2.

The list `wait_times` contains a list of seconds to wait. Write a program that reads a number from the list, then `sleep`s for that amount of time.

Since we do not want to wait too long, we won't sleep for any time longer than 1 second. If the wait time is longer than one second, skip straight to the next wait time.

Your program should announce what action it decides to take.

Example:

```
Waiting for 0.395 seconds...
Waiting for 0.596 seconds...
10.0 seconds is too long! Ain't nobody got time for that!
Waiting for 0.789 seconds...
```

Try to use the `continue` keyword since this is what we're working on.

> **NOTE:** if your program starts sleeping for a very long time, remember you can stop it with the stop button on the left side of the block.

````
# Another import from the standard library. The sleep function lets your program
# pause for an arbitrary numer of seconds. For example, sleep(1.0) stops for 1
# second before 
from time import sleep

print("Going to sleep...")
sleep(5.0)
print("I slept for 5 seconds!")
````

````
wait_times = [0.149, 0.624, 1.165, 0.128, 0.967, 1e6, 0.041, 12.536, 0.961, 0.591]

for t in wait_times:
  if t > 1.0:
    print(f"{t} seconds is too long! Ain't nobody got time for that!")
    continue
  print(f'Waiting for {t} seconds...')
  sleep(t)
````

## 3.

Write a 2-line program, using `zip` and `enumerate` that reads each character from 2 strings `a` and `b` and outputs a line containing the index `i`, the `i`th character of `a`, the `i`th character or `b`.

Example:

```python
a = 'Hello'
b = 'World'
```

Output:

```
0 H W
1 e o
2 l r
3 l l
4 o d
```

The program should stop when the shortest string is exhausted.

````
a = 'Whether I shall turn out to be the hero of my own life, or whether that station will be held by anybody else, these pages must show.'
b = 'It was the best of times, it was the worst of times, it was the age of wisdom, it was the age of foolishness, it was the epoch of belief, it was the epoch of incredulity, it was the season of Light, it was the season of Darkness, it was the spring of hope, it was the winter of despair.'

for i, (c1, c2) in enumerate(zip(a, b)):
  print(i, c1, c2)
````
