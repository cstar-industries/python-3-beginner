## 1.

The list `l` contains a number of arbitrary strings. Output the sum of all strings that represent an integer number.

Example:

```python
l = ['123', 'a31', '3b3', '321']
```

Output:

```
444
```

(444 = 123 + 321)

````
l = ['158','aLj','TGs','905','T0I','se2','20E','6ds','b6x','365','F4p','148','k7F','FVk','547']

total = 0
for s in l:
  if s.isnumeric():
    total += int(s)

print(total)
````

## 2. Set arithmetic

This exercise provides two 30-character strings `a` and `b`.

Use set arithmetic to print:

* All characters from either `a` or `b` (each character should only appear once)
* Only the characters that appear in `a` and `b` both
* The characters in `a` but not in `b`
* The characters in `b` but not in `a`

````
a = '13679BDGKLNRUVYZabcdefgijlqrsu'
b = '0134568DFHIMOPRWYZcijlmnopuwxz'

a = set(a)
b = set(b)

print(a | b)
print(a & b)
print(a - b)
print(b - a)
````

## 3. `set` inclusion performance

There is no problem to solve in this exercise. Just read the code in the blocks and run them one by one.

Which data structure is the more performant to test inclusion? Why do you think this is?

> **BONUS:** a few useful functions (like [`time`](https://docs.python.org/3/library/time.html#time.time) and [`randint`](https://docs.python.org/3.8/library/random.html#random.randint) are introduced in this section. Feel free to check out the docs and remember their usage.


````
# The `time` function returns the number of seconds elapsed since
# Jan 1 1970 00:00 UTC (UNIX Epoch). We'll be using it to time our experiments.
from time import time
````

````
# We will generate a set of 1 million random numbers between 0 and 2**64. Then
# we will get the numbers in that set as a list and compare the performance of
# both data structures.
from random import randint

s = set()

for _ in range(int(1e6)):
  s.add(randint(0, 1 << 64))

l = list(s)
print(f'Generated {len(s)} random numbers between 0 and {1 << 64}')
````

````
# First test: we go through every number from 0 to 999, and, if it is not in our
# numbers list, we add it to a nums_not_in_list list.
# WARNING: This is quite long, so be patient. The whole point of this exercise
# is to see how these data structure perform at scale.

nums_not_in_list = []

start_time = time()

for n in range(1000):
  if n not in l:
    nums_not_in_list.append(n)

end_time = time()

print(f'Found {len(nums_not_in_list)} numbers not in the list in {end_time - start_time:.2g} seconds')
````

````
# First test: we go through every number from 0 to 999, and, if it is not in our
# numbers set, we add it to a nums_not_in_set list.

nums_not_in_set = []

start_time = time()

for n in range(1000):
  if n not in s:
    nums_not_in_set.append(n)

end_time = time()

print(f'Found {len(nums_not_in_set)} numbers not in the set in {end_time - start_time:.2g} seconds')
````
