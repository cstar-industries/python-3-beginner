> Write a program that asks the user for an integer between 0 and 9 inclusive, then writes out the name of the digit in letters. If the number is not in the specified range, output an error message.
>
> Example:
> 
> ```
Enter an integer between 0 and 9: 4
four
Enter an integer between 0 and 9: 24
Sorry, number too large.
```

If you followed the rules back then, you probably wrote a lot of `if`s and `elif`s.

Write it againt this time, but use a `dict` to make the program simpler to read.

````
d = {0: 'zero',
     1: 'one',
     2: 'two',
     3: 'three',
     4: 'four',
     5: 'five',
     6: 'six',
     7: 'seven',
     8: 'eight',
     9: 'nine'}

x = int(input('Enter an integer between 0 and 9: '))

if x in d:
  print(d[x])
else:
  print('Sorry, number too large.')
````

> The following string is an encoded message. Each character in the message has been replaced by its 2-digit hexadecimal character number.
>
> $'A' \equiv 65_{decimal} \equiv 41_{hex}$

The code seemed repetitive back then. Rewrite it using `join` and a list comprehension to decode `hex_string` in just one line!

````
hex_string = '47726561742053636f747421'

''.join([chr(int(hex_string[i:i+2], 16)) for i in range(0, len(hex_string), 2)])
````
